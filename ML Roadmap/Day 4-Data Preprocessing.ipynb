```python
#Zscore normalization
# no library , no API
#scratch
#zscore= ( x- mean) / standard deviation

def zscore(df):  #function
    for col in df.columns:
        df[col]=(df[col]-df[col].mean())/df[col].std()
    return df

value=zscore(df[['HP','Att']])
value
```

```python
#Feature selection -> not null, only numeric
house=pd.read_csv('dataset/AmesHousing.csv')
house.isnull().sum()
house=house.drop('Lot Frontage',axis=1)
house.isnull().sum()
```

```python
len(house)
```

```python
cate_house=house.select_dtypes(include='object')
cate_house.head()
cate_house.isnull().sum()
```

```python
house=house.drop(columns=['Pool QC','Fence','Misc Feature','Alley'],axis=1)
```

```python
new_house=house.dropna()
new_house=new_house.select_dtypes(exclude='object')
new_house.head()
```

```python
#X,y ->feature, class label

X=new_house.drop('SalePrice',axis=1)
y=new_house['SalePrice']
```

```python
#Model ->
#Kbest  #chisquare  #LogisticRegression
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
x_new=SelectKBest(f_classif,k=5).fit_transform(X,y)
x_new
```

```python
x_new.shape
```

```python
from sklearn.feature_selection import chi2
x_new=SelectKBest(chi2,k=10).fit_transform(X,y)
x_new
```

```python
type(x_new) #n dimensional array
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
rfe=RFE(model)
values=rfe.fit(X,y)
```
