```python
from sklearn.impute import SimpleImputer
import numpy as np
imputer=SimpleImputer(missing_values=np.nan,strategy='mean') #only number "most-frequent"

#category - most_frequent
#number - mean, median, constant (e.g 100, 10)

values=data[['Att','Spd','Against Dark']]
result=np.round(imputer.fit_transform(values),2)
data[['Att','Spd','Against Dark']]=result
data
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent') #only number "most-frequent"
vv=data['Type 2'].values.reshape(-1,1)
result=imputer.fit_transform(vv)
data['Type 2']=result
data.head()
```

```python
data['Type 2'].values.reshape(-1,1) #1,-1
```

```python
from sklearn.impute import KNNImputer #Kth nearest neighbor #only number
imputer=KNNImputer(n_neighbors=5,weights="uniform")
vv=data[['Att','Spd','Against Dark']]
value=imputer.fit_transform(vv)
value
```
